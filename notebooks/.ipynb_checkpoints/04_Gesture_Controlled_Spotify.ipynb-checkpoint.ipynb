```python
import cv2
import joblib
import numpy as np
import pyautogui
import time

from skimage.feature import hog
```

```python
model = joblib.load("hand-gesture-recognition-model.sav")

print("Model Loaded Successfully!")
```

```python
labels = {
    0: "VolumeDown",
    1: "Previous",
    2: "Next",
    3: "PlayPause",
    4: "VolumeUp"
}
```

```python
display_text = {
    "PlayPause": "Play / Pause",
    "Next": "Next Track",
    "Previous": "Previous Track",
    "VolumeUp": "Volume Up",
    "VolumeDown": "Volume Down",
    "Place Hand": "Place Hand Inside Box"
}
```

```python
gesture_to_key = {
    "PlayPause": "playpause",
    "Next": "nexttrack",
    "Previous": "prevtrack",
    "VolumeUp": "volumeup",
    "VolumeDown": "volumedown"
}
```

```python
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Camera Error")
```

```python
previous_gesture = ""

gesture_counter = 0

required_frames = 5

last_action_time = time.time()

cooldown = 1.0
```

```python
while True:

    ret, frame = cap.read()

    if not ret:
        break

    frame = cv2.flip(frame,1)

    # -----------------------
    # ROI
    # -----------------------

    x1 = 350
    y1 = 100

    x2 = 600
    y2 = 350

    roi = frame[y1:y2, x1:x2]

    cv2.rectangle(frame,(x1,y1),(x2,y2),(0,255,0),2)

    # -----------------------
    # Preprocessing
    # -----------------------

    gray = cv2.cvtColor(roi,cv2.COLOR_BGR2GRAY)

    gray = cv2.resize(gray,(128,128))

    # -----------------------
    # Check ROI
    # -----------------------

    if gray.std() < 15:

        gesture = "Place Hand"

        gesture_counter = 0

        previous_gesture = ""

    else:

        feature_vector = hog(gray)

        feature_vector = feature_vector.reshape(1,-1)

        prediction = model.predict(feature_vector)

        gesture = labels[prediction[0]]

        # -----------------------
        # Consecutive Frames
        # -----------------------

        if gesture == previous_gesture:

            gesture_counter += 1

        else:

            previous_gesture = gesture

            gesture_counter = 1

        current_time = time.time()

        # -----------------------
        # Execute Action
        # -----------------------

        if gesture_counter >= required_frames:

            if (current_time-last_action_time) > cooldown:

                if gesture in gesture_to_key:

                    pyautogui.press(gesture_to_key[gesture])

                    last_action_time = current_time

                    gesture_counter = 0

    # -----------------------
    # Display
    # -----------------------

    action_text = display_text.get(gesture, gesture)

    cv2.putText(frame,
                action_text,
                (20,40),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.8,
                (0,255,0),
                2)

    cv2.putText(frame,
                f"Frames : {gesture_counter}",
                (20,80),
                cv2.FONT_HERSHEY_SIMPLEX,
                0.8,
                (255,0,0),
                2)

    cv2.imshow("AI Gesture-Controlled Spotify Media Controller", frame)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()

cv2.destroyAllWindows()
```
